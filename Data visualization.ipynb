```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x= np.array([0,7])
y=np.array([0,100])

plt.plot(x,y)
plt.show()
```

# Line Chart

```python
points = np.array([4,7,2,10,13])
plt.plot(points)
plt.show()
```

```python
n2 = np.array([2,4,8,6,10,20])
plt.plot(n2,linestyle="dotted")
plt.show()
```

# Bar Chart 

```python
x = np.array(["X","Y","Z","W"])
y = np.array([25,30,27,22])
plt.bar(x,y)
plt.show()
```

# Scatter Chart

```python
x1 = np.array([20,15,11,7,8,10])
y1 = np.array([2,7,15,8,10,12])
plt.scatter(x1,y1)
plt.show()
```

# Pie Chart

```python
p = np.array([30,10,22,44])
plt.pie(p)
plt.show()
```

## Histogram Chart

```python
h1 = np.random.normal(170,10,120)
plt.hist(h1)
plt.show()
```

## Dis Plot 

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
age1 = titanic['age'].dropna()
sns.displot(age1, bins=30 , kde=False) #dis give histogram for data 
```

## Reg Plot

```python
data = sns.load_dataset("mpg")
data.head()
```

```python
sns.regplot( x= 'mpg', y= 'acceleration', data=data) #Linear regression
plt.show()
```
